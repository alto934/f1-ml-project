```python
merge1 = pd.merge(races_df, weather_df, how='inner', on=['raceId','year', 'round', 'circuitId','name']).drop(['time','url','weather'],axis=1)
merge1.head()
```

```python
merge2 = pd.merge(merge1, results_df, how='left', on=['raceId']).drop(['number','positionText','positionOrder','fastestLap','rank','fastestLapTime','fastestLapSpeed','laps'],axis=1)

print(merge2.shape)
merge2.head()
```

```python
merge3 = pd.merge(merge2, drivers_df, how='left', on=['driverId']).drop(['number','code','url','forename','surname'],axis=1)

print(merge3.shape)
merge3.head()
```

```python
merge4 = pd.merge(merge3, circuits_df, how='left', on=['circuitId'], suffixes=('_mer3','_cir')).drop(['circuitId','location','lat',	'lng','url'],axis=1)

print(merge4.shape)
merge4 = merge4.drop(['name_cir'],axis=1)
merge4.head()
```

```python
merge5 = pd.merge(merge4, constructors_df, how='left', on=['constructorId'], suffixes=('_mer4','_con')).drop(['url'],axis=1)

print(merge5.shape)
merge5 = merge5.drop(['name','nationality_con'],axis=1)
merge5.head()
```

```python
print(merge5.columns)
```

```python
merge6 = pd.merge(merge5, qualifying_df, how='left', on=['constructorId','driverId','raceId'], suffixes=('_mer5','_qualy')).drop(['qualifyId','number','q1','q2','q3'],axis=1)

print(merge6.shape)
# merge5 = merge5.drop(['name','nationality_con'],axis=1)
merge6.head()
```

```python
print(merge6.columns)
```

```python
merge7 = pd.merge(merge6, driver_standings_df, how='left', on=['driverId','raceId'], suffixes=('_mer6','_driver')).drop(['positionText','driverStandingsId'],axis=1)

print(merge7.shape)
# merge5 = merge5.drop(['name','nationality_con'],axis=1)
merge7.head()
```

```python
print(merge7.columns)
```

```python
merge7.sort_values(['year','round'], inplace=True)
merge7.head()
```

```python

num_nans = merge7.isna().sum()

num_rows_with_nans = merge7.isna().any(axis=1).sum()
print(merge7.shape)
print(f"Number of NaN values in each column:\n{num_nans}\n")
print(f"Number of rows with NaN values: {num_rows_with_nans}")
merge7.to_csv('merge7.csv', index=False)
```

```python
merge8 = pd.merge(merge7, status_df , how='left', on=['statusId']).drop(['statusId'],axis=1)

print(merge8.shape)
# merge5 = merge5.drop(['name','nationality_con'],axis=1)
merge8.head()
```

```python
merge8.to_csv('merge8_nan.csv', index=False)
```

```python
merge8 = merge8.drop(['position_qualy', 'best_qualy_time'], axis=1)
```

```python

num_nans = merge8.isna().sum()

num_rows_with_nans = merge8.isna().any(axis=1).sum()
print(merge8.shape)
print(f"Number of NaN values in each column:\n{num_nans}\n")
print(f"Number of rows with NaN values: {num_rows_with_nans}")
```

```python
merge8 = merge8.fillna(0)
# merge8.dropna(inplace=True)
print(merge8.shape)

merge8.head()
```

```python
merge8 = merge8.rename(columns={'position_mer5': 'race_finish_pos', 'nationality_mer4':'nationality', 'position':'championship_pos','points_mer6':'points_awarded','wins':'cum_wins' })

```

```python
merge8 = merge8.drop(['time','milliseconds'],axis=1)
```

```python
merge8['race_finish_pos'] = merge8['race_finish_pos'].replace({'\\N':0})
merge8['race_finish_pos'] = pd.to_numeric(merge8['race_finish_pos'])
```

```python
#Change all values that have + so and so 'laps' at the end in the status column to 'Finished' 
merge8['status'] = merge8.status.str.replace("\+\d\s\w*|\+\d\d\s\w*", 'Finished')
```

```python
merge8['date'] = pd.to_datetime(merge8['date'])
merge8['dob'] = pd.to_datetime(merge8['dob'])

# difference between the two dates as a timedelta object
delta = merge8['date'] - merge8['dob']

# age at time of race in new column
merge8['driver_age'] = round(delta / pd.Timedelta(days=365.25),1)

merge8.head()
```

```python
print(merge8.shape)
```

```python
unique_nationality = merge8['nationality'].unique()
print(unique_nationality)
```

```python
unique_country = merge8['country'].unique()
print(unique_country)
```

```python
nationality_map = {
    'Finnish': 'Finland',
    'German': 'Germany',
    'British': 'UK',
    'Italian': 'Italy',
    'French': 'France',
    'Spanish': 'Spain',
    'Brazilian': 'Brazil',
    'Czech': 'Czech Republic',
    'Canadian': 'Canada',
    'Dutch': 'Netherlands',
    'Colombian': 'Colombia',
    'Japanese': 'Japan',
    'Australian': 'Australia',
    'Malaysian': 'Malaysia',
    'Irish': 'Ireland',
    'Danish': 'Denmark',
    'Hungarian': 'Hungary',
    'Austrian': 'Austria',
    'Indian': 'India',
    'Portuguese': 'Portugal',
    'American': 'USA',
    'Polish': 'Poland',
    'Swiss': 'Switzerland',
    'Russian': 'Russia',
    'Belgian': 'Belgium',
    'Venezuelan': 'Venezuela',
    'Mexican': 'Mexico',
    'Swedish': 'Sweden',
    'Indonesian': 'Indonesia',
    'New Zealander': 'New Zealand',
    'Monegasque': 'Monaco',
    'Thai': 'Thailand'
}

# create a new column called home_race
merge8['home_race'] = ( merge8['nationality'].map(nationality_map) == merge8['country']).astype(int)
merge8.head()
```

```python
merge8['points_before_race'] = merge8['points_driver'] - merge8['points_awarded']
merge8.head()
```

```python
merge8['wins'] = merge8['race_finish_pos'].apply(lambda x: 1 if x == 1 else 0)
merge8.head()
```

```python
print(merge8.columns)
```

```python
merge8 = merge8[['raceId','resultId','driverId','constructorId', 'year', 'round','name_mer3','nationality', 'circuitRef','driverRef', 'constructorRef','country','alt','date','dob', 'weather_warm','weather_cold', 'weather_dry', 'weather_wet', 
                 'weather_cloudy','grid', 'race_finish_pos','points_awarded',   'points_driver', 
                 'championship_pos','cum_wins', 'status', 'driver_age', 'home_race', 'points_before_race','wins']]
merge8 = merge8.drop(['alt'], axis=1)
```

```python
merge8['in_top_ten'] = merge8['race_finish_pos'].apply(lambda x: 1 if (x > 0 and x <= 10) else 0)
merge8['in_top_three'] = merge8['race_finish_pos'].apply(lambda x: 1 if (x > 0 and x <= 3) else 0)
```

```python
merge8.to_csv('merge8.csv', index=False)
```
